```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_html("https://en.wikipedia.org/wiki/List_of_European_Cup_and_UEFA_Champions_League_finals")
df
```

```python
df = pd.DataFrame(df[2])
pd.options.display.max_rows=100
df
```

```python
df = df.loc[:67, ['Season', 'Winners']].set_axis(labels=['Season', 'Nation', 'Team'], axis='columns')
pd.options.display.max_rows=100
df
```

```python
df.Season.duplicated()
```

```python
df = df.drop([19])
df
```

```python
# set new index
df['index'] = [x for x in range(len(df['Season']))]
df.set_index('index', inplace=True)
```

```python
season_list = []

for season in df['Season']:
  season_list.append(season[0:2]+season[-2:])

df['Season'] = season_list
df = df.drop(columns='Nation')
df
```

```python
df['win_default'] = 1
#df['check_rows'] = df['Team'] == (df['Team'].shift())
df
```

```python
df['cumulative'] = df[['Season', 'Team', 'win_default']].groupby('Team').cumsum() # Team & Season itu uniq. win_default utk cumulative sum
df
```

```python
df.info()
```

```python
most_win = df.groupby(['Team'])[['Season']].count().sort_values(by=['Season'], ascending=False)
most_win
```

```python
most_win = most_win.sort_values(by=['Season'], ascending=True)
most_win.plot(kind='barh')
```

```python
len(most_win)
```

```python
df_pivot = df.pivot_table(index='Season', columns='Team', values='cumulative')
df_pivot
```

```python
df_pivot = df_pivot.drop(['1900'])

df_pivot
```

```python
# fill first row with number of win, so i can fillna ffill
for club in df.Team:
  if club == 'Real Madrid':
    df_pivot.loc[df_pivot.index == '1956', 'Real Madrid'] = 1
  else:
    df_pivot.loc[df_pivot.index == '1956', :] = 0
  
df_pivot.head(2)
```

```python
df_pivot = df_pivot.fillna(method='ffill')
df_pivot = df_pivot.astype(int)

df_pivot
```

```python
df_pivot.to_csv('uclWInnersPivot.csv')
```
